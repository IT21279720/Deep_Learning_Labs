```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#Create a Random 4x4 Array
array_4x4 = np.random.exponential(scale=1.0, size=(4, 4))
print(array_4x4)
```

```python
#Create a 100,000x1 array from the same exponential distribution and visualize it.
exp_data = np.random.exponential(scale=1.0, size=(100000, 1))

uniform_data = np.random.uniform(low=-2.5, high=2.5, size=(100000, 1))

normal_data = np.random.normal(loc=0, scale=1, size=(100000, 1))

plt.hist(exp_data, density=True, bins=100, histtype="step", color="blue", label="Exponential")

plt.hist(uniform_data, density=True, bins=100, histtype="step", color="green", label="Uniform")

plt.hist(normal_data, density=True, bins=100, histtype="step", color="red", label="Normal")

plt.legend(loc="upper right")

plt.title("Random distributions")

plt.xlabel("Value")

plt.ylabel("Density")
```

```python
#Using matplotlib, plot the Z = X**2 + Y**2 in a 3D plot. Limit the x and y values to [-5, 5].
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
plt.show()
```

```python
# Check if all features exist in the DataFrame
missing_features = [feature for feature in features if feature not in df.columns]
if missing_features:
    print(f"Missing features: {missing_features}")
else:
    # Calculate Pearson correlation matrix
    pearson_correlation_matrix = df[features].corr(method='pearson')

    # Calculate Spearman correlation matrix
    spearman_correlation_matrix = df[features].corr(method='spearman')

    # Visualize the Pearson correlation matrix using a heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(pearson_correlation_matrix, annot=True, cmap='coolwarm')
    plt.title('Pearson Correlation Heatmap')
    plt.show()

    # Visualize the Spearman correlation matrix using a heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(spearman_correlation_matrix, annot=True, cmap='coolwarm')
    plt.title('Spearman Correlation Heatmap')
    plt.show()
```

```python
# Load the dataset

df = pd.read_csv('Au_nanoparticle_dataset.csv')

# Define the features

features = ['N_total', 'N_bulk', 'N_surface', 'R_avg']

# Calculate the Pearson and Spearman correlation

pearson_corr = df[features].corr(method='pearson')

spearman_corr = df[features].corr(method='spearman')

# Visualize the results using heatmaps

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)

sns.heatmap(pearson_corr, annot=True, cmap='coolwarm')

plt.title('Pearson Correlation')

plt.subplot(1, 2, 2)

sns.heatmap(spearman_corr, annot=True, cmap='coolwarm')

plt.title('Spearman Correlation')

plt.show()
```

```python
#Create a new dataframe by filtering all the columns [i.e., features] except N_total, N_bulk, N_surface and R_avg columns.
df = pd.read_csv('Au_nanoparticle_dataset.csv')
filtered_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

```python
#Display the first 20 samples of this dataframe.
print(filtered_df.head(20))
```

```python
#Calculate the mean, standard deviation and quartile values for each of the above 4 features.
stats = filtered_df.describe()
mean_values = filtered_df.mean()
std_values = filtered_df.std()
quartiles = filtered_df.quantile([0.25, 0.5, 0.75])
print("Means:\n", mean_values)
print("Standard Deviations:\n", std_values)
print("Quartiles:\n", quartiles)
```

```python
#Plot the histogram of each of these features in a 1x4 layout.
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
filtered_df.hist(column=['N_total', 'N_bulk', 'N_surface', 'R_avg'], ax=axes)
plt.show()
```

```python
#Visualize the scatter plots and histograms of this dataframe using the pairplot functionality of seaborn library.
sns.pairplot(filtered_df)
plt.show()
```

```python
#Plots on the diagonal contains the histogram of each feature along with the kernal density estimation plot.
g = sns.PairGrid(filtered_df)
g.map_upper(sns.histplot)
g.map_diag(sns.histplot, kde=True)
g.map_lower(sns.kdeplot)
plt.show()
```
